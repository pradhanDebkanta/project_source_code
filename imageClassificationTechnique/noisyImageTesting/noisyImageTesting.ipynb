```python
path_normal = '/content/drive/MyDrive/kidneyDataSet/test_speckle_noise_4/Normal/'
path_stone = '/content/drive/MyDrive/kidneyDataSet/test_speckle_noise_4/Stone/'
path_cyst = '/content/drive/MyDrive/kidneyDataSet/test_speckle_noise_4/Cyst/'
path_tumor = '/content/drive/MyDrive/kidneyDataSet/test_speckle_noise_4/Tumor/'
```

```python
def img_prediction(img):
    image = cv2.imread(img)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(gray_image, (200, 200), interpolation=cv2.INTER_AREA)
    y = np.expand_dims(resized, axis=-1)
    y = np.expand_dims(y, axis=0)

    automl = keras.models.load_model('/content/drive/MyDrive/kidneyStoneModel/model_CNN.h5')

    predictions = automl.predict(y).flatten()
    return predictions
```

```python
{'Cyst': 0, 'Normal': 1, 'Stone': 2, 'Tumor': 3}
```

```python
# img_class='Tumor'
classes = ['Cyst','Normal','Stone','Tumor']

cyst_accuracy=0
normal_accuracy=0
stone_accuracy=0
tumor_accuracy=0

for img_class in classes:
  if img_class=='Cyst':
      print('Cyst class test')

      images = os.listdir(path_cyst)
      total_cyst_img_count=len(images)
      total_prediction_img_count = 0
      
      for idx, img in enumerate(images):
          get_prediction = img_prediction(path_cyst+'{}'.format(img))
          # print('predicted result==>', get_prediction)
          if int(get_prediction[0]) == 1:
              total_prediction_img_count+=1
      print('total_prediction_img_count==>', total_prediction_img_count)
      print("len==>",total_cyst_img_count )
      cyst_accuracy=(total_prediction_img_count/total_cyst_img_count)*100
      
  elif img_class == 'Normal':
      print('Normal class test')

      images = os.listdir(path_normal)
      total_normal_img_count=len(images)
      total_prediction_img_count = 0
      
      for idx, img in enumerate(images):
          get_prediction = img_prediction(path_normal+'{}'.format(img))
          # print('predicted result==>', get_prediction)
          if int(get_prediction[1]) == 1:
              total_prediction_img_count+=1
      print('total_prediction_img_count==>', total_prediction_img_count)
      print("len==>",total_normal_img_count )
      normal_accuracy=(total_prediction_img_count/total_normal_img_count)*100

  elif img_class=='Stone':
      print('Stone class test')

      images = os.listdir(path_stone)
      total_stone_img_count=len(images)
      total_prediction_img_count = 0
      
      for idx, img in enumerate(images):
          get_prediction = img_prediction(path_stone+'{}'.format(img))
          # print('predicted result==>', get_prediction)
          if int(get_prediction[2]) == 1:
              total_prediction_img_count+=1
      print('total_prediction_img_count==>', total_prediction_img_count)
      print("len==>",total_stone_img_count )
      stone_accuracy=(total_prediction_img_count/total_stone_img_count)*100


  elif img_class=='Tumor':
      print('Tumor class test')

      images = os.listdir(path_tumor)
      total_tumor_img_count=len(images)
      total_prediction_img_count = 0
      
      for idx, img in enumerate(images):
          get_prediction = img_prediction(path_tumor+'{}'.format(img))
          # print('predicted result==>', get_prediction)
          if int(get_prediction[3]) == 1:
              total_prediction_img_count+=1
      print('total_prediction_img_count==>', total_prediction_img_count)
      print("len==>",total_tumor_img_count )
      tumor_accuracy=(total_prediction_img_count/total_tumor_img_count)*100
      

  else :
      print('no Class found')
```

```python
print("Cyst img accuracy==>", cyst_accuracy)
print("Normal img accuracy==>", normal_accuracy)
print("Stone img accuracy==>", stone_accuracy)
print("Tumor img accuracy==>", tumor_accuracy)
```

```python
# calculate avg accuracy

overall_accuracy=(cyst_accuracy + normal_accuracy + stone_accuracy + tumor_accuracy)/4

print('overall accuracy==>', overall_accuracy)
```
